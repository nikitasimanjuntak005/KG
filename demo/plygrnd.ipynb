```python
import numpy as np

path_pos = '../data/train_pos.npy'
path_neg = '../data/train_neg.npy'
```

```python
try:
    # Load data from file .npy
    data_pos = np.load(path_pos)
    data_neg = np.load(path_neg)

    # --- Positive Data Information ---
    print("Positive Data (train_pos.npy):")
    # Display 5 first row
    print(data_pos[:5]) 
    # Displat data dimension (row and col count)
    print("Positive Data Dimension:", data_pos.shape)
    print("-" * 30)

    # --- Negative Data Information ---
    print("Negative Data (train_neg.npy):")
    # Display 5 first row
    print(data_neg[:5])
    # Display data dimension
    print("Negative Data Dimension:", data_neg.shape)

except FileNotFoundError:
    print("Error: File not found.")
```
